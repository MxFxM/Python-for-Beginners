```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 10, 0.2)
y = np.sin(x)

# the paper on which you are plotting
fig = plt.figure()

# a graph on the paper
# a figure can have multiple axes on it
ax = fig.add_subplot(1, 1, 1) # matlab starts at 1
ax.plot(x, y)

plt.savefig("07 pyplot1.png", dpi=300)
```

```python
fig = plt.figure()
fig.set_size_inches(8.5, 4)
```

```python
ax1 = fig.add_axes((0.1, 0.2, 0.7, 0.5)) # left botttom width height
ax1.plot(x, y)

ax2 = fig.add_axes((0.9, 0.9, 0.1, 0.1))
ax2.plot(x, y)

fig
```

```python
fig, axes = plt.subplots(1, 3, figsize=(12,4))
for ax in axes:
    ax.plot(x, y)
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1) # matlab starts at 1
ax.plot(x, y, color='r', linewidth=5, linestyle='--')

plt.savefig("07 pyplot2.png", dpi=300)
```

Plot Types

```python
fig, ax = plt.subplots()
ax.bar(x, y)
```

```python
fig, ax = plt.subplots()
ax.scatter(x, y, marker='x')
```

```python
fig, ax = plt.subplots()
ax.scatter(x, y, c=y, cmap='viridis')
# c is the color, here based on the y value
# cmap is the color map
```

```python
fig, ax = plt.subplots()
ax.boxplot(y)
```

```python
fig, ax = plt.subplots()
ax.violinplot(x)
```

```python
fig, ax = plt.subplots()
ax.violinplot([x, y])
```

```python
fig, ax = plt.subplots()

ax.hist(y)
```

```python
fig, ax = plt.subplots()
ax.pie([10, 5], labels=["10", "5"])
ax.legend()
```
